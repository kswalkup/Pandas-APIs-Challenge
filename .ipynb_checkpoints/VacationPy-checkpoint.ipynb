### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_csv_file = "output_data/weather_data.csv"
weather_df = pd.read_csv(weather_csv_file)
weather_df.dropna()
weather_df.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = weather_df[["Latitude", "Longitude"]]
humidity = weather_df["Humidity"].astype(float)
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=max(humidity),
                                 point_radius=3.5)

fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
idealWeather_df = weather_df.loc[(weather_df['Cloudiness'] < 25) &
                                 (weather_df['Humidity'] < 50) & 
                                 (weather_df['Max Temperature'] < 80) &
                                 (weather_df['Max Temperature'] > 50) &
                                 (weather_df['Wind Speed'] < 15)]

idealWeather_df = idealWeather_df.dropna()
idealWeather_df.head()
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# Code from "NearestRestr" heavily referenced to complete the hotel map section, and some hints from Bob
hotel_df = idealWeather_df

#params dictionary to update each iteration
params = {"radius": 5000,
    "type": "lodging",
    "key": g_key}

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():

#get coordinates from df
    lat = row["Latitude"]
    lng = row["Longitude"]

# add keyword to params dict
    params["location"] = f"{lat},{lng}"
    
# assemble url and make API request
    print(f"Retrieving Results for Index {index}: {row['City']}.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"{results[0]['name']} is the nearest hotel.")
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```

```python
hotel_df = hotel_df.dropna()
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Latitude", "Longitude"]]
```

```python
hotelList = hotel_df["Hotel Name"].tolist()
```

```python
# Referenced "Census_States" exercise to create this code

locations = weather_df[["Latitude", "Longitude"]]
humidity = weather_df["Humidity"].astype(float)
marker_locations = hotel_df[['Latitude', 'Longitude']]

fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=max(humidity),
                                 point_radius=3.5)

markers = gmaps.marker_layer(marker_locations,
    info_box_content=[f"Acceptable Hotels: {list}" for list in hotelList])

fig.add_layer(heat_layer)
fig.add_layer(markers)
fig
```
